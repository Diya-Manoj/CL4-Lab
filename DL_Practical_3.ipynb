```python
import numpy as np
import re
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

### Load Dataset

```python
vocab_size = 10000
max_len = 500

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)
```

### Padding

```python
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)
```

### Model

```python
model = Sequential()
model.add(Embedding(vocab_size, 128, input_length=max_len))
model.add(LSTM(64))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Train

```python
model.fit(X_train, y_train, epochs=3, batch_size=64, validation_split=0.2)
```

### Evaluate

```python
model.evaluate(X_test, y_test)
```

### Preprocessing

```python
word_index = imdb.get_word_index()

def preprocess_review(text):
    text = re.sub(r"[^a-zA-Z ]", "", text)
    words = text.lower().split()
    
    encoded = [1]
    
    for word in words:
        if word in word_index and word_index[word] < vocab_size:
            encoded.append(word_index[word] + 3)
        else:
            encoded.append(2)
    
    padded = pad_sequences([encoded], maxlen=max_len)
    return padded
```

### Prediction Function

```python
def predict_sentiment(text):
    processed = preprocess_review(text)
    pred = model.predict(processed)
    return "Positive" if pred[0][0] > 0.5 else "Negative"
```

### Manual Testing

```python
reviews = [
    "The movie was fantastic and full of emotions",
    "Worst movie ever waste of time"
]

for r in reviews:
    print(r)
    print("Sentiment:", predict_sentiment(r))
    print()
```

### User Input

```python
user_review = input("Enter a movie review:\n").strip()

print("Sentiment:", predict_sentiment(user_review))
```

📌 Code Overview: Sentiment Analysis using LSTM (RNN Variant)

This implementation focuses on building a sentiment analysis model using a Recurrent Neural Network (RNN) variant — Long Short-Term Memory (LSTM) on the IMDb movie reviews dataset.

🔹 1. Dataset Loading

The IMDb dataset is loaded using TensorFlow/Keras. It contains 50,000 movie reviews labeled as positive (1) or negative (0).
To limit complexity, only the top 10,000 most frequent words are considered.

🔹 2. Data Preprocessing
Reviews are already numerically encoded in the dataset.
All sequences are padded to a fixed length using pad_sequences() to ensure uniform input size for the model.
For custom input, text is:
Cleaned (removing special characters)
Converted to lowercase
Tokenized into words
Encoded using IMDb word index with proper offset
Padded to match training input size
🔹 3. Model Architecture

A Sequential Deep Learning model is built using:

Embedding Layer → Converts word indices into dense vector representations
LSTM Layer → Captures sequential dependencies and context in text
Dense Layer (Sigmoid Activation) → Outputs probability for binary classification

The model is compiled using:

Loss Function: Binary Crossentropy
Optimizer: Adam
Metric: Accuracy
🔹 4. Model Training

The model is trained on the training dataset using:

Multiple epochs
Batch processing
Validation split to monitor performance
🔹 5. Model Evaluation

The trained model is evaluated on test data to measure its generalization ability using accuracy and loss.

🔹 6. Prediction Pipeline

A unified preprocessing function is created to ensure consistent transformation of input data.
A reusable prediction function is implemented to:

Accept raw text input
Preprocess it
Generate prediction using the trained model
Classify sentiment as Positive or Negative
🔹 7. Testing
Predictions are performed on:
Predefined sample reviews
User-input reviews for real-time sentiment analysis
🎯 Key Concept

The LSTM network processes text sequentially and retains contextual information using memory cells, making it more effective than traditional RNNs for handling long-term dependencies in language data.

🧠 One-Line Summary

👉 Text → Preprocess → Encode → Pad → LSTM → Predict Sentiment
